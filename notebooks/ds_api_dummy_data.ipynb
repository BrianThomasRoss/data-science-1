## Make function that combines dataframes and uses model to make a prediction

```
df = pd.read_csv('https://raw.githubusercontent.com/fp-rap-build/data-science/main/data/dummy_data.csv')

df
```

## Hard Encode Input Dictionary

```
input = {
    'zipcode': 99001,
    'monthly_income': 5000,
    'family_size': 4,
    'unemployed_90+_days': 1,
    'eviction_notice': 1,
}
```

```
def framemaker(web_in):
# making dataframe out of dict  
  input_frame = pd.DataFrame(web_in, index=[0])

  input_frame = input_frame[['zipcode', 'monthly_income', 'family_size', 'unemployed_90+_days', 'eviction_notice']]

  userinput = input_frame.iloc[[0]]

  return userinput 
```

```
user_input = framemaker(input)
```

```
import pandas as pd
from sklearn.model_selection import train_test_split
import category_encoders as ce
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegressionCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from category_encoders import OneHotEncoder, OrdinalEncoder
!pip install category_encoders==2.*
```

```
def model_maker():
  train, test = train_test_split(df, train_size=0.80, test_size=0.20, 
                                 stratify=df['qualify_for_snap'], random_state=42)

  # Arrange data into X features matrix and y target vector
  features = ['zipcode', 'monthly_income', 'family_size', 'unemployed_90+_days', 'eviction_notice']
  target = ['qualify_for_snap']

  X_train = train[features]
  y_train = train[target]
  X_test = test[features]
  y_test = test[target]

  lrmodel = Pipeline([
                  ('ohe', OneHotEncoder(use_cat_names=True)),
                  ('scaler', StandardScaler()),  
                  ('impute', SimpleImputer()),
                  ('classifier', LogisticRegressionCV())
                  ])
  lrmodel.fit(X_train, y_train)

  return lrmodel

lrmodel = model_maker()

def predict(user_input):

  if lrmodel.predict(user_input) == 1:
    predict = {'predict': 'Applicant qualifies for SNAP!'}
    input.update(predict) 
    return input
  else:
    predict = {'predict':'Applicant qualifies for Family Promise!'}
    input.update(predict)
    return input
```
