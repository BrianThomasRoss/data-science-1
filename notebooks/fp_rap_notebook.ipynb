```python
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/fp-rap-build/data-science/main/data/zipcode_raw%20-%20data.csv')

data.head()
```

```python
df = data[['name', 'city_name', 'median_household_income', 'average_household_income']].copy()

df.head()
```

```python
df = df.rename(columns={'name':'zipcode',})
```

```python
df.head()
```

```python
df['monthly_median_income'] = df['median_household_income'].div(12).round(2)
```

```python
df.head()
```

```python
df['monthly_average_income'] = df['average_household_income'].div(12).round(2)
```

```python
df.head()
```

```python
df_small = df[['zipcode', 'city_name', 'monthly_median_income', 'monthly_average_income']].copy()
```
